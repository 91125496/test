# matploilib

```python
# importation de package
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generation de Données
x = np.linspace(0,10,1000)
y = np.exp(x)
```

```python
# Graphique Simple
plt.figure()
plt.plot(x,y,)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,c='green', ls='-',label='exp')
plt.subplot(2,2,2)
plt.plot(x,y,c='blue', ls='-',label='exp')
plt.subplot(2,2,3)
plt.plot(x,y,c='black', ls='-',label='exp')
plt.subplot(2,2,4)
plt.plot(x,y,c='red', ls='-',label='exp')
plt.subplots_adjust(wspace=0.5)
plt.savefig('figure.png')
```

```python
plt.plot(x,y,c='green', ls='--',label='exp')
plt.title('fonction exponentiel')
plt.xlim(6,10)
plt.ylim(1000,20000)
plt.xlabel('X')
plt.ylabel('exp')
plt.legend()
```
